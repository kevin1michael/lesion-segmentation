```python
from PIL import Image

im = Image.open("./out/ISIC_0000000_2.png")

pix = im.load()

width, height = im.size

pixel_array = []
pixel_row = []
region_labels = set([0])
current_region = 0

isNewRegion = True
foundAllRegions = False

while not foundAllRegions:
    for y in range(height):
        for x in range(width):
            pixel_value = pix[x, y]
            if pixel_value not in region_labels and isNewRegion:
                current_region = pixel_value
                isNewRegion = False
                region_labels.add(pixel_value)
            
            if pixel_value == current_region:
                pixel_row.append(pixel_value)
            else:
                pixel_row.append(0)

        pixel_array.append(pixel_row)
        pixel_row = []

    if isNewRegion:
        foundAllRegions = True
    else:
        isNewRegion = True
        X = np.array(pixel_array, np.int32)
        pixel_array = []
        print_props(X)
```
